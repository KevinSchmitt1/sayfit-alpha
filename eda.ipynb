```python
df_usda_copy = df_usda_clean.copy()
df_usda_copy = df_usda_copy.drop(columns=['item_id', 'brand',])

df_off_copy = df_off.copy()
df_off_copy = df_off_copy.drop(columns=['item_id', 'brand', 'serving_size', 'quantity'])
```

```python
df_full = pd.concat([df_usda_copy, df_off_copy], ignore_index=True)
```
